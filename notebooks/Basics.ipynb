```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
```

```python
exchange_rates.head()
```

```python
exchange_rates.tail()
```

```python
exchange_rates.info()
```

### Data Cleaning

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
```

```python
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
```

```python
exchange_rates.sort_values('Time', inplace=True)
```

```python
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
exchange_rates
```

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']]
euro_to_dollar
```

```python
euro_to_dollar['US_dollar'].value_counts()
```

```python
euro_to_dollar = euro_to_dollar[~(euro_to_dollar['US_dollar'] == '-')]
euro_to_dollar['US_dollar'].value_counts()
```

```python
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype('float')
euro_to_dollar.info()
```

### Generating plots

```python
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.show()
```

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar.plot('Time', 'rolling_mean')
plt.show()
```

### Graphical Analysis of Euro-Dollar rate changed during the 2007-2008 financial crisis

We show how the euro-dollar rate changed during the 2007-2008 financial crisis. We can also show the data for 2006 and 2009 for comparison. We can use a line plot.

### Starting with graph 1

```python
# importing the styles 
import matplotlib.style as style
```

```python
# Defining the graphical style
style.use('fivethirtyeight')
```

```python
# Selecting the date interval to plot
select_range = euro_to_dollar[euro_to_dollar['Time'].between('2006-01-04', '2011-01-04')]
# Selecting the beginning of 2007 and end of 2008
crisis_range = euro_to_dollar[euro_to_dollar['Time'].between('2007-01-01', '2008-12-31')]

# Implementing the graph
fig, ax = plt.subplots(figsize=(9,5))
# Creating the reference line plot
ax.plot(select_range['Time'], select_range['rolling_mean'],
        color = '#00FF00', linewidth = 0.5)
# Creating the crisis range line plot
ax.plot(crisis_range['Time'], crisis_range['rolling_mean'],
       color = '#af0b1e', linewidth = 4)

# Changing the y ticks
ax.set_yticks([1.2, 1.3, 1.4, 1.5])

# Adding span during the euro-dollar rate in 1.5
ax.axvspan(xmin=733120.0, xmax=733300.0, ymin=0.09,
           alpha=0.3, color='grey')
    

# Adding the title
ax.text(732100, 1.63, "Euro-USD peaked at 1.59 during 2007-2008's financial crisis",
       size = 16, weight = 'bold')
# Adding the subtitle
ax.text(732100, 1.60, "Euro-USD exchange rates between 2006 and 2010")
# Plotting the graph
# Adding the watermark
plt.show()
```
